## Atualizar a versao das imagens

```python
import ee
from pprint import pprint
```

```python
ee.Initialize()
```

```python
asset = "projects/nexgenmap/MapBiomas2/PLANET/images"
```

```python
collection = ee.ImageCollection(asset)\
    .filter(ee.Filter.inList('tile', [
        '1633613',#ok
        '1633559',#ok
        '1633410',#ok
        '1347079',#ok
        '1633922',#ok
        '1634947',#ok
        '1635782',#ok
        '1731555',#ok
        '1731730',#ok
        '1746515',#ok
        # '1635541'
    ]))
```

```python
versions = collection.distinct("version")\
    .reduceColumns(ee.Reducer.toList(), ["version"])\
    .get('list')\
    .getInfo()

print(versions)
```

```python
count = 1
for version in versions:
    
    subcollection = collection.filterMetadata('version', 'equals', version)
    
    imageNames = subcollection\
        .reduceColumns(ee.Reducer.toList(), ['system:index'])\
        .get('list')\
        .getInfo()

    size = subcollection.size().getInfo()

    for imageName in imageNames:
        
        inputName = asset +'/'+ imageName
        outputName = asset +'/'+ imageName[:-2]
        
        print('[{}/{}] replacing for version {} {}'.format(count, size, version, outputName))

        ee.data.copyAsset(
            sourceId=inputName, 
            destinationId=outputName, 
            allowOverwrite=True)
        
        ee.data.setAssetProperties(outputName, {"version": None})
        
        count = count + 1

```
